```python
import sys
from bs4 import BeautifulSoup
import requests
import json
import urllib.request
import csv
```

```python
q1 = {'country':'IN',
      'date_from':'2015-01-01',
      'date_to':'2019-09-16',
      'limit':10000,
      'format':'csv',
     }
base_url = 'https://api.openaq.org/v1/measurements'
```

```python
r = requests.get(base_url, params = q1)
r.url
```
