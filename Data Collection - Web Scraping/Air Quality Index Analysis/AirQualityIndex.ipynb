```python
mydata = pd.read_csv('raw_data.csv')
```

```python
mydata = mydata[mydata['city'] != 'city'] 
```

```python
#Drop unnecessary columns
df = mydata.drop(columns=["attribution","utc"])
```

```python
df.dtypes
```

```python
#Extracting date from field - local
df["date"] = df["local"].astype(str).str.slice(0,10) 
```

```python
#converting datatype of column "value" to numeric
df["value"] = pd.to_numeric(df["value"])
```

```python
#Grouping the dataset on all columns except "value" and taking average of "value"
df = df.groupby(['location','city','country','latitude','longitude','unit','date','parameter'], as_index=False)['value'].mean()
```

```python
df.head()
```

```python
#Creating new columns by pivot
df_pivot = pd.pivot_table(df, values = 'value', index=['location','city','country','unit','latitude','longitude','date'], columns = 'parameter').reset_index()
```

```python
df_pivot.head(10)
```

```python
#Handling missing values - replacing null values with mean
df_pivot = df_pivot.fillna(df_pivot.mean())
```

```python
df_pivot.to_csv('air_quality_index_pivot.csv')
```
