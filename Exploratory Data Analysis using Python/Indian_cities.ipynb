```python
#We first import the packages that we would need for this case study
%matplotlib inline
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
import numpy as np
```

```python
#Read the csv File and get total number of rows and coloumns
Indian_cities=pd.read_csv('C:\\Users\\ditsaxen\\Desktop\\Personal\\ISB CBA\\Session Material and Projects\\Projects\\Python project\\Indian_cities.csv')
Indian_cities.shape
```

```python
#getting the data type of each variable
Indian_cities.dtypes
```

```python
#View the firts few rows of the data set
Indian_cities.head(10)
```

```python
#get summary of data
Indian_cities.describe().apply(lambda s:s.apply(lambda x:format(x,'g')))
```

```python
#Dropping few coloumn values to make data set compact , containing only required fields
Indian_cities=Indian_cities.drop('dist_code',1)
Indian_cities=Indian_cities.drop('state_code',1)
Indian_cities=Indian_cities.drop('location',1)
```

```python
#dispalying the filtered data set
Indian_cities
```

```python
#Filtering out unique state names which will play major role in the analysis of the data
States=Indian_cities['state_name'].unique()
print(States)
print(len(States))
```

```python
#Calculating total population for each state
Indian_cities.groupby('state_name')['population_total'].sum()
```

```python
#plot graph for State Vs Total population
df=pd.DataFrame(Indian_cities.groupby('state_name')['population_total'].sum())
df.plot(kind='bar',stacked=True,figsize=(15,8))
```

```python
#plotting graph for total males and population through out all the states
Totalmales=Indian_cities['population_male'].sum()
Totalfemales=Indian_cities['population_female'].sum()
df=pd.DataFrame([Totalmales,Totalfemales])
df.index=['Males','Females']
df.plot(kind='bar',stacked=True,figsize=(20,8))
```

```python
#plotting graph for males vs females in each state
df=pd.DataFrame(Indian_cities.groupby('state_name')['population_male','population_female'].sum())
df.plot(kind='bar',stacked=True,figsize=(20,10))
```

```python
#plotting graph for total kids under 6(male,female) for each state
df=pd.DataFrame(Indian_cities.groupby('state_name')['0-6_population_male','0-6_population_female'].sum())
df.plot(kind='bar',stacked=True, color= ['g','b'],figsize=(20,10))
```

```python
#Plotting grpah to compare overall literacy rateof each state
df=pd.DataFrame(Indian_cities.groupby('state_name')['effective_literacy_rate_total'].sum())
df.plot(kind='bar',stacked=True,figsize=(20,10))
```

```python
#Plotting graph for literacy rate among male and female in each state
df=pd.DataFrame(Indian_cities.groupby('state_name')['literates_male','literates_female'].sum())
df.plot(kind='bar',stacked=True, color= ['g','b'],figsize=(20,10))
```

```python
#Plotting graph for sex ratio and child sexratio in each state
df=pd.DataFrame(Indian_cities.groupby('state_name')['sex_ratio','child_sex_ratio'].sum())
df.plot(kind='bar',stacked=True,figsize=(20,10))
```

```python
df=pd.DataFrame(Indian_cities.groupby('state_name')['effective_literacy_rate_male','effective_literacy_rate_female'].sum())
df.plot(kind='bar',stacked=True,figsize=(20,10))
```

```python
#Graph to display overall literacy rate, male literacy and female literacy rate, all together for each state
df=pd.DataFrame(Indian_cities.groupby('state_name')['effective_literacy_rate_total','effective_literacy_rate_male','effective_literacy_rate_female'].sum())
df.plot(kind='bar',stacked=True,figsize=(20,10))
```

```python
#plottng graph to check no of graduates(male,female) in each state to compare
df=pd.DataFrame(Indian_cities.groupby('state_name')['total_graduates','male_graduates','female_graduates'].sum())
df.plot(kind='bar',stacked=True,figsize=(20,10))
```
