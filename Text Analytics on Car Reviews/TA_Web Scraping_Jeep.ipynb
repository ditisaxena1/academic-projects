# Kia Seltos reviews from "www.zigwheels.com"

```python
import sys
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
```

```python
page = requests.get("https://www.zigwheels.com/user-reviews/Kia/seltos#leadform").text
soup_obj = BeautifulSoup(page, "lxml")
```

```python
reviews_title = soup_obj.find_all('div', class_ = "f-rv-des mb-10 clr-bl fnt-16 read-more")
reviews_title = [n.text for n in reviews_title]

review_rating = soup_obj.find_all('span',class_ =['r-n g' , 'r-n lg'])
review_rating = [n.text for n in review_rating]

review_date = soup_obj.find_all('span', class_ ='ur-ncd fnt-14 clr-sry')
review_date = [n.string for n in review_date]
```

```python
#Removing special characters and expressions
review_date = [x.strip(' ') for x in review_date]
review_date = [x.replace('\r', '') for x in review_date]
review_date = [x.replace('\n', '') for x in review_date]

review_rating = [x.strip(' ') for x in review_rating]

reviews_title = [x.replace('\r', '') for x in reviews_title]
reviews_title = [x.replace('\n', '') for x in reviews_title]
reviews_title = [x.replace('...', '') for x in reviews_title]
reviews_title = [x.replace('?', '') for x in reviews_title]
reviews_title = [x.replace(';', '') for x in reviews_title]
reviews_title = [x.replace(')', '') for x in reviews_title]
reviews_title = [x.replace('..', '') for x in reviews_title]
reviews_title = [x.strip(' ') for x in reviews_title]
```

```python
kia = pd.DataFrame( {'Review':reviews_title, 'Review Date':review_date, 'Star Ratings':review_rating} )
kia
```

```python
for n in kia['Review']:
    kia['Review Website']='www.zigwheels.com'
    kia['Brand Name']='Kia Seltos'   
    
kia
```

# Jeep Compass reviews from "www.cars.com" 720 rows

```python
import sys
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
jeepc_reviewdate =list()
jeepc_reviewstitle1 =list()
jeepc_reviewstitle =list()
JeepReviews=list()
JeepReviews1=list()
JeepRevDate=list()
jeepreviewdate=list()
jeepcreview1=list()

for i in range (1, 37):
    url = "https://www.cars.com/research/jeep-compass-2018/consumer-reviews/?pg="+str(i)+"&nr=10"
    
    page = requests.get(url).text
    soup_obj=BeautifulSoup(page,"lxml")
        
    jeepc_reviews_title = soup_obj.find_all('p', class_ = "cui-heading-6")
    jeepcreview = [n.text for n in jeepc_reviews_title]
    JeepReviews.extend(jeepcreview)
    
    jeepc_reviews_title1 = soup_obj.find_all('p', class_ = "review-card-text")
    jeepcreview1 = [n.text for n in jeepc_reviews_title1]
    JeepReviews1.extend(jeepcreview1)
      
    jeepc_review_date = soup_obj.find_all('p', class_ ='review-card-review-by')
    jeepreviewdate = [n.text for n in jeepc_review_date]
    JeepRevDate.extend(jeepreviewdate)
```

```python
jeepc = pd.DataFrame( {'Review':JeepReviews, 'Review Date':JeepRevDate} )
jeepc1 = pd.DataFrame( {'Review':JeepReviews1, 'Review Date':JeepRevDate} )
df3= jeepc.append(jeepc1)
for n in df3['Review']:
    df3['Review Website']='www.cars.com'
    df3['Brand Name']='Jeep Compass'   
    df3['Star Ratings']=5    
```

```python
df3['Review'] = [x.replace('\r', '') for x in df3['Review']]
df3['Review'] = [x.replace('\n', '') for x in df3['Review']]
df3['Review'] = [x.replace('...', '') for x in df3['Review']]
df3['Review'] = [x.replace('?', '') for x in df3['Review']]
df3['Review'] = [x.replace(';', '') for x in df3['Review']]
df3['Review'] = [x.replace(')', '') for x in df3['Review']]
df3['Review'] = [x.replace('..', '') for x in df3['Review']]
df3['Review'] = [x.strip(' ') for x in df3['Review']]

df3['Review Date'] = [x.strip(' ') for x in df3['Review Date']]
df3['Review Date'] = [x.replace('\r', '') for x in df3['Review Date']]
df3['Review Date'] = [x.replace('\n', '') for x in df3['Review Date']]
```

```python
df4 = df2.append(df3)
df4
```
